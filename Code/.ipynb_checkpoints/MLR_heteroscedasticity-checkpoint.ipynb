```python
import matplotlib.pyplot as plt 
import pandas as pd
import numpy as np
import statsmodels.api as sm
import statsmodels.formula.api as smf
```

```python
credit = pd.read_csv("Credit.csv")
credit['Income'] = pd.to_numeric(credit['Income'])
```

```python
# fit initial model after taking care of multicollinearity
model =smf.ols('Balance~Income + Rating +Cards+Age +Education +Gender+Student+Married+Ethnicity',data=credit).fit()
model.summary()
```

```python
#Heteroscedasticity
#Fitted Values vs. Residuals Plot
p = model.fittedvalues
res = model.resid
plt.scatter(p,res)
plt.xlabel("Fitted Values")
plt.ylabel("Residual")
plt.title("Fitted Values vs. Residuals")
#looks bad
#the observations around y=0 is causing a lot of variaty in this data
```

```python
#Breusch-Pagan for Heteroskedasticity
from statsmodels.stats.diagnostic import het_breuschpagan
bp_test = het_breuschpagan(model.resid, model.model.exog)
labels = ['LM Statistic', 'LM-Test p-value']
print(dict(zip(labels, bp_test)))
# Heteroskedasticity exists
```

```python
#white test
#from statsmodels.stats.diagnostic import het_white
#white_test = het_white(reg.resid,  reg.model.exog)
```

```python
#ln transformation on y
# This residul plot doesn't change proportionally with hat_y, so ln transformation on y is not gonna work well
credit['log_balance']=np.log(credit['Balance'])

print(credit['log_balance'])
# ln transofrmation doesn't work on 0 or negative values
# there are a lot of discussion around this, but a quick solution would be adding a constant
```

```python
# to deal with 0 values, add a small number to y: 
credit['log_balance_1']=np.log(credit['Balance']+1)
```

```python
model_1 =smf.ols('log_balance_1~Income + Rating +Cards+Age +Education +Gender+Student+Married+Ethnicity',data=credit).fit()
```

```python
p1 = model_1.fittedvalues
res1 = model_1.resid
plt.scatter(p1,res1)
plt.xlabel("Fitted Values")
plt.ylabel("Residual")
plt.title("Fitted Values vs. Residuals after log transformation on y")
```

Clearly, the y=0 observations behave very differently from other observations, might suggest a analysi only on observations with non-zero y values.

```python
# another example
df = pd.read_csv('kc_house_data.csv')
df['log_price'] = np.log(df['price'])
```

```python
f = 'price~sqft_living'
model2 = smf.ols(formula=f, data=df).fit()
p2 = model2.fittedvalues
res2 = model2.resid
plt.scatter(p2,res2)
plt.xlabel("Fitted Values")
plt.ylabel("Residual")
plt.title("Fitted Values vs. Residuals")
```

```python
f1 = 'log_price~sqft_living'
model3 = smf.ols(formula=f1, data=df).fit()
p3 = model3.fittedvalues
res3 = model3.resid
plt.scatter(p3,res3)
plt.xlabel("Fitted Values")
plt.ylabel("Residual")
plt.title("Fitted Values vs. Residuals")

#the heteroscedasticity is improved, but of course this data comes with more problems such as influential points
```

<b>Example of solution: robust standard errors </b>
    
While it is common to use robust standard errors in economics, the documentation in Python’s StatsModels is not as rich as expected. We can grab the robust standard errors, or Heteroskedasticity-Consistent Standard Errors (HCE) using some built-in functions.

```python
#orginal model that we know having heteroskedasticity problem
model =smf.ols('Balance~Income + Rating +Cards+Age +Education +Gender+Student+Married+Ethnicity',data=credit).fit()

#extact HCE variance-coviance matrix
V_HC_1 =model.cov_HC0
print(V_HC_1)
```

```python
#While we don't care to see the actully corrected se values, we can extract the corrected t test and p values directly
print(np.round(model.get_robustcov_results(cov_type = "HC0").summary2().tables[1], 5))
```

```python
#Compare to the orginal model summary, numbers are slightly changed.
model.summary()
```

```python
# weighted least squares: optional
plt.figure(figsize=(20, 3))

# i: index
for i, col in enumerate(credit.columns):
    plt.subplot(1, 14, i+1)
    x = credit[col]
    y = model.resid
    plt.plot(x, y, 'o')

    plt.title(col)
    plt.xlabel(col)
    plt.ylabel('residuls')
    
    # looks like residuals are proportional to Income and Rating
    
    
```

```python
credit["res_sq"] = model.resid**2
model_res = smf.ols("res_sq ~ Income+Rating", data=credit).fit()  
weight = model_res.fittedvalues
weight = abs(weight) 
weight = weight**-1 
credit['weight'] = weight

```

```python
y=credit.Balance
x=credit[['Income', 'Rating','Cards','Age','Education']] 

```

```python

mod_wls = sm.WLS(y,x.astype(float), credit['weight'])
mod_res = mod_wls.fit()
print(mod_res.summary())
```
