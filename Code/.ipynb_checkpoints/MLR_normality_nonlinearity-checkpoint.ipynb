```python
import matplotlib.pyplot as plt 
import pandas as pd
import numpy as np
import statsmodels.api as sm
import statsmodels.formula.api as smf
```

```python
credit = pd.read_csv("Credit.csv")
credit['Income'] = pd.to_numeric(credit['Income'])
```

```python
model =smf.ols('Balance~Income + Rating +Cards+Age +Education +Gender+Student+Married+Ethnicity',data=credit).fit()
model.summary()
```

```python
#histogram of residual to observe the shape/skewness
plt.hist(model.resid)
```

```python
#QQ plot
sm.qqplot(model.resid)
```

```python
from scipy.stats import kstest
kstest(model.resid,'norm')
```

```python
credit['log_balance_1']=np.log(credit['Balance']+1)
model_1 =smf.ols('log_balance_1~Income + Rating +Cards+Age +Education +Gender+Student+Married+Ethnicity',data=credit).fit()
model_1.summary()
```

```python
# Nonlinearity beween y and x
#pair-wise scatter plot between y and predictors
plt.figure(figsize=(20, 3))


# i: index
for i, col in enumerate(credit.columns):
    plt.subplot(1, 14, i+1)
    x = credit[col]
    y = credit['Balance']
    plt.plot(x, y, 'o')

    plt.title(col)
    plt.xlabel(col)
    plt.ylabel('Blance')
    # looks okay: maybe income and rating?
```

```python
#residual plot might have suggestions too
#pair-wise scatter plot between residuals and predictors
plt.figure(figsize=(20, 3))


# i: index
for i, col in enumerate(credit.columns):
    plt.subplot(1, 14, i+1)
    x = credit[col]
    y = model.resid
    plt.plot(x, y, 'o')

    plt.title(col)
    plt.xlabel(col)
    plt.ylabel('residuls')
```

```python
credit['log_rating']=np.log(credit['Rating'])
credit['log_income']=np.log(credit['Income'])
```

```python
model_2 =smf.ols('log_balance_1~log_income + log_rating +Cards+Age +Education +Gender+Student+Married+Ethnicity',data=credit).fit()
model_2.summary()
# with the same number of predictors, R^2 improved
```

```python
#another example 
df = pd.read_csv('kc_house_data.csv')
df.sample(5)
```

```python
predictor=df[['sqft_living', 'sqft_lot','sqft_basement' ]]
for i, col in enumerate(predictor.columns):
    plt.subplot(1, 3, i+1)
    x = predictor[col]
    y = df.price
    plt.plot(x, y, 'o')

    plt.title(col)
    plt.xlabel(col)
    plt.ylabel('y')
```

```python
df['log_price'] = np.log(df['price'])
df['log_sqft_living'] = np.log(df['sqft_living'])
df['log_sqft_lot'] = np.log(df['sqft_lot'])
df['log_sqft_basement'] = np.log(df['sqft_basement'])
```

```python
predictor=df[['log_sqft_living', 'log_sqft_lot','log_sqft_basement' ]]
for i, col in enumerate(predictor.columns):
    plt.subplot(1, 3, i+1)
    x = predictor[col]
    y = df.log_price
    plt.plot(x, y, 'o')

    plt.title(col)
    plt.xlabel(col)
    plt.ylabel('lny')
```

Generally speaking, normality and nonlinearity are harder to improve, but the same time a smaller problem that we need to worry about.We can always try log transformations and scale the predictors, that should be sufficient.
